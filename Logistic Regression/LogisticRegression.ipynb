```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Generate 2D random data
X = np.random.rand(100, 2)
y = np.random.randint(0, 2, 100)

# Train model
lg = LogisticRegression()
lg.fit(X, y)

# Create a grid over the feature space
x1_range = np.linspace(X[:,0].min(), X[:,0].max(), 100)
x2_range = np.linspace(X[:,1].min(), X[:,1].max(), 100)
xx1, xx2 = np.meshgrid(x1_range, x2_range)
grid = np.c_[xx1.ravel(), xx2.ravel()]

# Get predicted probabilities for the grid
probs = lg.predict_proba(grid)[:, 1].reshape(xx1.shape)

# Plot
plt.figure(figsize=(7,5))
plt.contourf(xx1, xx2, probs, cmap='coolwarm', alpha=0.7)
plt.scatter(X[:,0], X[:,1], c=y, edgecolor='k', cmap='coolwarm')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Surface')
plt.colorbar(label='Predicted Probability (Class 1)')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

X = np.random.rand(100, 3)
y = np.random.randint(0, 2, 100)

lg = LogisticRegression()
lg.fit(X, y)

# For visualization, fix one feature (say X3 = mean)
x1_range = np.linspace(X[:,0].min(), X[:,0].max(), 50)
x2_range = np.linspace(X[:,1].min(), X[:,1].max(), 50)
x3_fixed = np.mean(X[:,2])

xx1, xx2 = np.meshgrid(x1_range, x2_range)
grid = np.c_[xx1.ravel(), xx2.ravel(), np.full(xx1.ravel().shape, x3_fixed)]

probs = lg.predict_proba(grid)[:,1].reshape(xx1.shape)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx1, xx2, probs, cmap='coolwarm', alpha=0.8)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Predicted Probability')
plt.title('3D Logistic Regression Surface (Feature3 fixed)')
plt.show()
```
