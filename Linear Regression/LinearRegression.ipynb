```python
import pandas as pd
import numpy as np
```

Implementing Linear regression from sratch

```python
# reproducible integer samples between 1 and 99 (20 values)
# np.random.seed(0)
X = np.random.randint(1, 100, 20).astype(float)
y = np.random.randint(1, 100, 20).astype(float)
```

The Best fit Line is given by the ( y^p = bo + b1*x )

<img src="image.png" width="300">

```python
b1=(len(X)*(np.sum(X*y))-np.sum(X)*np.sum(y))/(len(X)*np.sum(X**2)-np.sum(X)**2)
bo=(np.sum(y)-b1*np.sum(X))/len(X)

print(b1)
print(bo)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, bo + b1*X, color='red', label='Best Fit Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression from Scratch')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(X.reshape(-1, 1), y)

import matplotlib.pyplot as plt
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, model.predict(X.reshape(-1, 1)), color='green', label='Sklearn Best Fit Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Sklearn')
```

```python

print("Sklearn slope:\t\t", model.coef_[0])
print("From scratch slope:\t", b1)

print("Sklearn intercept:\t", model.intercept_)
print("From scratch intercept:\t", bo)
```
